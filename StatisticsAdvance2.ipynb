Question 1: What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population, based on a sample of data.
    - Hypotheses
    - Test Statistic
    - P-value
    - Significance Level

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - Hypothesis testing involves two competing statements: the null hypothesis and the alternative hypothesis.

  **Difference:**
  
- Null Hypothesis
    - It is the default assumption that there is no effect, no difference, or no relationship in the population.

    - It represents the existing belief or status quo.

    - It usually contains an equality sign

- Alternative Hypothesis

  - It is the statement that contradicts H₀ and represents what the researcher wants to prove.

  - It suggests that there is an effect, difference, or relationship.

  - It uses symbols like **not-equal, less than, greater than**

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance levelin hypothesis testing is the probability of rejecting the null hypothesis when it is actually true. It represents the researcher’s tolerance for making a Type I error.

**Role in Decision Making:**
- After calculating the p-value (probability of observing the data under H₀).

- The significance level defines the critical region (rejection region) in the probability distribution of the test statistic.

Question 4: What are Type I and Type II errors? Give examples of each.
- **Type I Error** : Rejecting the null hypothesis when it is actually true.

  - Example: Convicting an innocent person in a court trial.

- **Type II Error**: Failing to reject the null hypothesis when it is actually false.

  - Example: Acquitting a guilty person in a court trial.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
- **Z-test:**

  - Used when the population standard deviation (σ) is known.

  - Applicable for large sample sizes (n ≥ 30).

  - Test statistic follows the standard normal (Z) distribution.

  - Example: Testing if the mean height of adults differs from 170 cm when σ is known.

- **T-test:**

  - Used when the population standard deviation (σ) is unknown and is estimated from the sample.

  - Suitable for small sample sizes (n < 30).

  - Test statistic follows the Student’s t-distribution.

  - Example: Testing if the mean weight of a small sample of students differs from 50 kg when σ is unknown.

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 10000

data = np.random.binomial(n, p, size)


plt.hist(data, bins=range(n+2), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50

n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

se = sample_std / np.sqrt(n)

z_stat = (sample_mean - mu_0) / se


p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.4f}")
print(f"Sample Std Dev = {sample_std:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
mu, sigma = 50, 5
n = 100

data = np.random.normal(mu, sigma, n)


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

confidence = 0.95
alpha = 1 - confidence
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)

ci_lower = sample_mean - t_critical * se
ci_upper = sample_mean + t_critical * se

print(f"Sample Mean = {sample_mean:.4f}")
print(f"95% Confidence Interval = ({ci_lower:.4f}, {ci_upper:.4f})")


plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")

plt.title("Normal Distribution Data with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = (data - mean) / std_dev
    return z_scores


data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                 50.3, 50.4, 50.0, 49.7, 50.5, 49.9])

z_scores = calculate_z_scores(data)


plt.hist(z_scores, bins=10, color='lightblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (0)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


print("First 10 Z-scores:", z_scores[:10])
```
